```python
import collections
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv as csv
import shutil
import os
import json as json
import tensorflow as tf
```

<h3>Converting json data to csv</h3>

```python
outfile = open("review_stars.tsv", 'w')
sfile = csv.writer(outfile, delimiter ="\t", quoting=csv.QUOTE_MINIMAL) 
sfile.writerow(['business_id','stars', 'text'])
with open('./yelp_dataset/yelp_academic_dataset_review.json', encoding='utf-8') as f: 
    for line in f:
        row = json.loads(line)
        # some special char must be encoded in 'utf-8' 
        sfile.writerow([row['business_id'], row['stars'], (row['text']).encode('utf-8')])
outfile.close()
```

```python
outfile2 = open("business_dataset.tsv", 'w')
sfile2 = csv.writer(outfile2, delimiter ="\t", quoting=csv.QUOTE_MINIMAL) 
sfile2.writerow(['business_id','review_count', 'stars'])
with open('./yelp_dataset/yelp_academic_dataset_business.json', encoding='utf-8') as f2: 
    for line in f2:
        row = json.loads(line)
        # some special char must be encoded in 'utf-8' 
        sfile2.writerow([row['business_id'], row['review_count'],  row['stars']])
outfile2.close()
```

<h3>Reading file into data frame</h3>

```python
df1 = pd.read_csv('review_stars.tsv', delimiter ="\t", encoding="utf-8",na_values=['NA','?'])
df2 = pd.read_csv('business_dataset.tsv', delimiter ="\t", encoding="utf-8",na_values=['NA','?'])
```

```python
dfa = df1[0:50000]
dfb = df2[0:50000]
dfa
```

```python
dfb
```

```python
review_agg = dfa.groupby('business_id')['text'].sum()
df_review_agg = review_agg.to_frame().reset_index()
df_review_agg[0:10]
```

```python
df_merged = pd.merge(df_review_agg, dfb, on="business_id")
df_merged[0:10]
```

```python
df_merged.shape
```

<h3>Label Encoding</h3>

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
encoded_star = le.fit_transform(df_merged['stars'])
encoded_star[0:50]
```

```python
import sklearn.feature_extraction.text as sk_text
# vectorizer = sk_text.CountVectorizer(min_df=1, stop_words = 'english')
vectorizer = sk_text.TfidfVectorizer(
                             stop_words='english',
                             max_features = 1000,
                             min_df=2, max_df=0.5)
matrix = vectorizer.fit_transform(df_merged.text)
matrix.shape
```

```python
tfidf_data = matrix.toarray()
print(tfidf_data)
```

<b> Normalizing data using zscore </b>

```python
from scipy.stats import zscore

df_merged['normalize_rc'] = zscore(df_merged['review_count'])
df_merged[0:5]
```

```python
y = df_merged['normalize_rc'].values.reshape( len(df_merged['normalize_rc']),1)
tfidf_data = np.concatenate([tfidf_data,y],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(tfidf_data,encoded_star, test_size=0.20, random_state=42)
print(tfidf_data.shape)
print(encoded_star.shape)
```

<h2>Regression</h2>

```python
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard, EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Activation, Dense
from keras.utils.np_utils import to_categorical
import matplotlib.pyplot as plt
model = Sequential()

model.add(Dense(25, input_dim=tfidf_data.shape[1], activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')  

monitor = EarlyStopping(monitor='val_loss', min_delta=1e-4, patience=3, verbose=2, mode='auto') 

checkpointer = ModelCheckpoint(filepath="best_weights.hdf5", verbose=0, save_best_only=True) # save best model
model.fit(x_train, y_train, validation_data=(x_test,y_test), callbacks=[monitor,checkpointer], verbose=2, epochs=1000)  
model.load_weights('best_weights.hdf5')
```

```python
pred = model.predict(x_test)
from sklearn import metrics
score = np.sqrt(metrics.mean_squared_error(pred,y_test))
print("Final score (RMSE): {}".format(score))
```

<h3>OneHotEncoding</h3>

```python
import collections
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import shutil
import os


# Encode text values to dummy variables(i.e. [1,0,0],[0,1,0],[0,0,1] for red,green,blue)
def encode_text_dummy(df, name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = "{}-{}".format(name, x)
        df[dummy_name] = dummies[x]
    df.drop(name, axis=1, inplace=True)
```

```python
df_encoded_star = pd.DataFrame(encoded_star, columns=['one_hot_encoded'])
encode_text_dummy(df_encoded_star, 'one_hot_encoded')
df_encoded_star[0:5]
```

```python
one_hot_array=df_encoded_star.values
one_hot_array[:10]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(tfidf_data,one_hot_array, test_size=0.20, random_state=42)
```

```python
model = Sequential()
model.add(Dense(25, input_dim= tfidf_data.shape[1], activation='sigmoid')) # Hidden 1
model.add(Dense(10, activation='sigmoid')) # Hidden 2
model.add(Dense(one_hot_array.shape[1], activation='softmax')) # Output

model.compile(loss='categorical_crossentropy', optimizer='rmsprop')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-4, patience=3, verbose=2, mode='auto') 

checkpointer = ModelCheckpoint(filepath="best_weights.hdf5", verbose=0, save_best_only=True) # save best model

model.fit(x_train, y_train, validation_data=(x_test,y_test), callbacks=[monitor,checkpointer], verbose=2,epochs=1000)
```

```python
pred_classification = model.predict(x_test)
from sklearn import metrics
score = np.sqrt(metrics.mean_squared_error(pred_classification,y_test))
print("Final score (RMSE): {}".format(score))
```

```python
# Measure accuracy
pred = model.predict(x_test)
pred = np.argmax(pred,axis=1)

y_true = np.argmax(y_test,axis=1)

score = metrics.accuracy_score(y_true, pred)
print("Final accuracy: {}".format(score))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(tfidf_data,encoded_star, test_size=0.20, random_state=42)
print(tfidf_data.shape)
print(encoded_star.shape)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(x_train, y_train)

# predict the response for new observations
Log_pred=logreg.predict(x_test)

score = np.sqrt(metrics.mean_squared_error(Log_pred,y_test))
print("Final score (RMSE): {}".format(score))

# Measure accuracy
score = metrics.accuracy_score(y_test, Log_pred)
print("Final accuracy: {}".format(score))
```

```python
#Support Vector Machine
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train, y_train)  
svm_pred=clf.predict(x_test)

score = np.sqrt(metrics.mean_squared_error(svm_pred,y_test))
print("Final score (RMSE): {}".format(score))

# Measure accuracy
score = metrics.accuracy_score(y_test, svm_pred)
print("Final accuracy: {}".format(score))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
print(knn)
knn.fit(x_train, y_train)
Knn_pred=knn.predict(x_test)  

score = np.sqrt(metrics.mean_squared_error(Knn_pred,y_test))
print("Final score (RMSE): {}".format(score))

# Measure accuracy
score = metrics.accuracy_score(y_test, Knn_pred)
print("Final accuracy: {}".format(score))
```

```python
# Multinomial Naive Bayes
from sklearn import preprocessing
from sklearn.naive_bayes import MultinomialNB
min_max_scaler = preprocessing.MinMaxScaler()
mnb = MultinomialNB()
mnb.fit(min_max_scaler.fit_transform(x_train), y_train)
mnb_pred = mnb.predict(min_max_scaler.fit_transform(x_test))

score = np.sqrt(metrics.mean_squared_error(mnb_pred,y_test))
print("Final score (RMSE): {}".format(score))

# Measure accuracy
score = metrics.accuracy_score(y_test, mnb_pred)
print("Final accuracy: {}".format(score))
```
